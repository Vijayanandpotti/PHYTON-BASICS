### The follwoing are the bit wise operators




### `COMPLEMENT    ~`

- #### Complement means it will do reverse of the binary format.

- ####  it means if the result is 0, complement will give you the the output as 1.

- #### in virtual memory it cannot store negative number, that is why we use complement.


```python
print(bin(12))
print(bin(13))
```

- #### bin means binary.

### `LEFT SHIFT <<`

1. Left shift means we gain the bit 

```python
bin(10)
```

```python
10<<1 # this code refers  to 1 bit (after symbol)
```

```python
10<<2  # this code refers to 2 bit (after symbol)
```

###  `RIGHT SHIFT >>`

- ####  Right shift means we loose the bit 

```python
bin(10)
```

```python
10>>2 # this code refers to 2 bit (after symbol)
```
